## IFNULL

IFNULL returns the first of the two indicated values if the data value found in
the table is not null, and returns the second value if there is a null
value.
Same as coalesce but limited with 2 arguements

## COALESCE

1. think of COALESCE() as IFNULL() with more than two parameters
2. COALESCE() will always return a single value of the ones we have 
within parentheses, and this value will be the first non-null value of 
this list, reading the values from left to right

```python
%%sql
SELECT dept_no,
COALESCE(dept_name,'Department name not provided') as dept_name
FROM departments_dup;
```

```python
%%sql
SELECT dept_no,
COALESCE('department_manager_name') as fake_col
FROM departments_dup;
```

## HAVING clause

refines the output that do not satisfy a certain condition.
1. HAVING can have condition with an aggreagate fuction
2. WHERE cant have condition with an aggregate function

```python
%%sql
SELECT * FROM employees
HAVING hire_date >= "1995-01-01"
LIMIT 30
```

```python
%%sql
SELECT * FROM employees
HAVING COUNT(first_name) > 250
ORDER BY first_name;
```

Extracting list of all names that are encountered greater than 1 times. Let the data refer to the people hired after the 1st of january 1995

```python
%%sql
SELECT first_name, COUNT(first_name) AS names_count FROM employees
WHERE hire_date > '1995-01-01'
GROUP BY first_name
HAVING COUNT(first_name) >1
ORDER BY names_count;
```

## LIMIT

## DROP vs TRUNCATE vs DELETE

1. When DROP is used you won’t be able to roll back to its initial state, or to the last COMMIT statement
2. use DROP TABLE only when you are sure you aren’t going to use the table in question anymore
3. When truncating, auto-increment values will be reset
4. Delete - removes records row by row

## INNER JOIN

Its similar to A intersection B

FORMAT OF THE CODE;

SELECT
t1.column_name, t1.column_name, …, t2.column_name, …

FROM 
table_1 t1
JOIN table_2 t2 ON t1.column_name = t2.column_name;

```python
%%sql
SELECT * FROM dept_manager_dup
ORDER BY dept_no;
```

```python
%%sql
SELECT * FROM departments_dup
ORDER BY dept_no;
```

```python
%%sql
SELECT m.dept_no, m.emp_no,d.dept_name
FROM dept_manager_dup m
INNER JOIN departments_dup d ON m.dept_no = m.dept_no
ORDER BY m.dept_no;
```

## JOIN

JOIN removes the duplicate values

```python
%%sql
INSERT INTO dept_manager_dup
VALUES ('110228', 'd003', '1992-03-21','9999-01-01');

INSERT INTO departments_dup
VALUES ('d009', 'Customer Service');
```

```python
%%sql
SELECT * FROM dept_manager_dup
ORDER BY dept_no ASC;
```

To remove the duplicate value of 110228

```python
%%sql
SELECT m.dept_no, m.emp_no, d.dept_name
FROM dept_manager_dup m
JOIN departments_dup d ON m.dept_no = d.dept_no
ORDER BY dept_no;
```

## LEFT JOIN

Similar to (A-B) operation in matrx

```python
%%sql
DELETE FROM dept_manager_dup
WHERE emp_no = '110228';

DELETE FROM departments_dup
WHERE dept_no = "d009";

INSERT INTO dept_manager_dup
VALUES('110228','d003','1992-03-21','9999-01-01');

INSERT INTO departments_dup
VALUES ('d009','Customer Service')
```

```python
%%sql
SELECT m.dept_no, m.emp_no, d.dept_name
FROM dept_manager_dup m
JOIN departments_dup d ON m.dept_no = d.dept_no
GROUP BY m.emp_no
ORDER BY m.dept_no;
```

## CROSS JOIN

1. connects all the values, not just those that match
2. particularly useful when the tables in a database are not well connected

```python
%%sql
SELECT dm.*,d.* FROM dept_manager dm
CROSS JOIN departments d
ORDER BY dm.emp_no , d.dept_no;
```
