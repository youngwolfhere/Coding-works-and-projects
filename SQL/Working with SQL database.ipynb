## HAVING clause

refines the output that do not satisfy a certain condition.

Extracting list of all names that are encountered greater than 1 times. Let the data refer to the people hired after the 1st of january 1995

```python
%%sql
SELECT first_name, COUNT(first_name) AS names_count FROM employees
WHERE hire_date > '1995-01-01'
GROUP BY first_name
HAVING COUNT(first_name) >1
ORDER BY names_count;
```

## LEFT JOIN

The Left Join in SQL basically returns all records from the left table and the matched records from the right tables

The Left Join in SQL basically returns all records from the left table and the matched records from the right tables

```python
%%sql
DELETE FROM dept_manager_dup
WHERE emp_no = '110228';

DELETE FROM departments_dup
WHERE dept_no = "d009";

INSERT INTO dept_manager_dup
VALUES('110228','d003','1992-03-21','9999-01-01');

INSERT INTO departments_dup
VALUES ('d009','Customer Service')
```

## CROSS JOIN

1. connects all the values, not just those that match. IN other words, the cartesian product of the values of two or more sets.
2. particularly useful when the tables in a database are not well connected

```python
%%sql
SELECT dm.*,d.* FROM dept_manager dm
CROSS JOIN departments d
ORDER BY dm.emp_no , d.dept_no;
```

or

```python
%%sql
SELECT dm.* , d.* FROM dept_manager dm, departments d
ORDER BY dm.emp_no, d.dept_no;
```

or

Even if we use inner join or join instead of cross join without explicitly stating the condition, sql will implement the cross join

```python
%%sql
SELECT dm.*,d.* FROM dept_manager dm
JOIN departments d
ORDER BY dm.emp_no , d.dept_no;
```

To remove the repeated values of dept no from dept_no_1 column. In other words, we dont want the department to be shown for where the manager is working

```python
%%sql
SELECT dm.*,d.* FROM departments d
CROSS JOIN dept_manager dm
WHERE d.dept_no <> dm.dept_no
ORDER BY dm.emp_no , d.dept_no;
```

## Cross join on more than two tables

```python
%%sql
SELECT e.*,d.* FROM departments d
CROSS JOIN dept_manager dm
JOIN employees e ON dm.emp_no =e.emp_no
WHERE d.dept_no <> dm.dept_no
ORDER BY dm.emp_no , d.dept_no;
```

Assigning employee number 110022 as a manager to all employees from 10001 to 10020 and employee number 110039 as a manager to all employees from 10021 to 10040

```python
%%sql
SELECT A.* FROM (SELECT e.emp_no AS employee_ID,
        MIN(de.dept_no) AS department_code,
        (SELECT emp_no FROM dept_manager Where emp_no = 110022 ) AS manager_ID
        FROM employees e 
        JOIN dept_emp de ON e.emp_no = de.emp_no
        WHERE e.emp_no <= 10020
        GROUP BY e.emp_no
        ORDER BY e.emp_no) as A
UNION
SELECT B.* FROM (SELECT e.emp_no AS employee_ID,
        MIN(de.dept_no) AS department_code,
        (SELECT emp_no FROM dept_manager Where emp_no = 110039) AS manager_ID
        FROM employees e 
        JOIN dept_emp de ON e.emp_no = de.emp_no
        WHERE e.emp_no > 10020 AND e.emp_no <=10040
        GROUP BY e.emp_no
        ORDER BY e.emp_no) as B ; 
```
